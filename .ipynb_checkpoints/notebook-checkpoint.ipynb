# Microsoft Film Analysis

### Overview

This project is to help determine what kind(s) of films Microsoft should make as they start their new film division. Information was collected from IMDb, Box Office Mojo, and The Number on films released since 2019 as well as actors based both in the US and abroad. Microsoft can use this information to decide what movies they should make.

### Business Problem

Microsoft can choose what films to make as well as how to ensure that their movies are profitable by selecting a reasonable production budget, making films that tend to be profitable, and hiring lead actors whose movies tend to perform well. Using data from the websites mentioned above, I discover what a reasonable budget for a major motion picture is, which genres of film produce the higher return on investment, and which actors have the most successful films so that Microsoft can hire them for a lead role. 

### The Data

First, we will import all of the libraries and tools we need in order to effectively break apart and analyze the data.

```python
import pandas as pd
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Now, we will take a look at some basic information about modern movies, provided by IMDb. This information includes the title, year of release, run time, and which genre(s) it was categorized as.

```python
title_basics = pd.read_csv('data/imdb.title.basics.csv.gz')

title_basics
```

While taking a preliminary look at the available data, I found that I could join some useful information, but in order to do that, I need to rename the "primary_title" column to be called "movie".

```python
title_basics.columns = ['tconst','movie','original_title','start_year','runtime_minutes','genres']

title_basics
```

Now we can look at some information about production budgets and gross revenues. This information came from The Numbers.

```python
budgets = pd.read_csv('data/tn.movie_budgets.csv.gz')

budgets
```

Here, I combine the previous two tables together so that I can do some calculations later on.

```python
query = """

SELECT *
FROM title_basics
JOIN budgets
USING (movie)

"""

results = pysqldf(query)

results
```

I wanted to make sure that movies in this chart were recent, so that revenue numbers could be comparable to today's standards. Revenue from a movie that was made in, say, the 1970s may be difficult to compare to revenue in today's dollars.

```python
results['start_year'].describe()
```

Next, I had to format the budget and revenue columns so that they were floats that I could do math with, instead of strings. While the dollar signs and commas are useful for reading, they're not very useful for my purposes.

```python
results_clean = results[['movie', 'genres']]
results_clean
```

```python
prod_budgets = results['production_budget']

prod_budgets
```

```python
prod_costs = prod_budgets.apply(lambda x: float(x.replace("$","").replace(",","")))

prod_costs
```

```python
world_gross = results['worldwide_gross']

world_gross
```

```python
gross_revenue = world_gross.apply(lambda x: float(x.replace("$","").replace(",","")))

gross_revenue
```

```python
dom_rev = results['domestic_gross']

dom_rev
```

```python
domestic_gross = dom_rev.apply(lambda x: float(x.replace("$","").replace(",","")))

domestic_gross
```

Now, I can add these newly formatted arrays back into the table.

```python
final_table = results_clean.join(prod_costs, how='inner').join(domestic_gross, how='inner').join(gross_revenue, how = 'inner')

final_table
```

Then, I wanted to know the return on investment, or roi, of each movie to determine how profitable a given movie is.

A high roi can also be a proxy for popularity. Since revenue is from ticket sales, lots of ticket sales mean that lots of people  wanted to see it.

```python
net_revenue = final_table['worldwide_gross'] - final_table['production_budget']
final_table['roi'] = net_revenue / final_table['production_budget']

final_table
```

Now that I have the complete table that I want to use, I can look at some information that I think will be helpful, starting with budgets. 

```python
final_table['production_budget'].median()
```

```python
final_table['production_budget'].mean()
```

I see that the average is much higher than the median, meaning the data is skewed toward movies with much higher budgets. I think this makes a lot of sense: not all movies are super high budgets, but the ones made by major film studios like Disney, Sony and the like will be much more expensive than everyone else. Microsoft's budget should be within the average of major studios, since those will be Microsoft's competitors, rathern than smaller or niche film studios with much lower budgets. 

But a budget is only half the equation: the amount of money going into a film. Let's see how budgets relate to revenue numbers.

```python
final_table_query1 = """

SELECT *
FROM final_table
WHERE production_budget > 32000000 
ORDER BY worldwide_gross

"""

results1 = pysqldf(final_table_query1)

results1
```

```python
results1.plot('production_budget','worldwide_gross', kind='scatter')
```

It looks lke there is some sort of positive relationship between production budget and revenue. Further down we can look at some hard numbers (after making a properly formatted graph).

```python
results1_prod = results1['production_budget']
results1_gross = results1['worldwide_gross']

plt.scatter(results1_prod,results1_gross)
plt.title('Production Budgets vs Gross Revenue')
plt.xlabel('Production Budget in Hundreds of Millions')
plt.ylabel('Gross Revenue in Billions')
```

```python
results1.corr()
```

It looks like there is a strong positive correlation between budget and gross revenue, with an r = 0.709. So we can confidently say that the more mmoney you spend, the more money you'll likely make. But, production budget has an almost non-existent relationship with roi.

So, we will have to find a way to ensure a higher roi. I figured one way we could ensure roi is by choosing what genre of film to make, so let's take a look.

```python
final_table['genres'].describe()
```

So we have lots of different combinations of genres, 345 to be exact. Well it's not very useful to have 345 groups to compare. So, I'm going to split them up so that each movie has only one primary genre. I'm going to asssume that the first genre list in a given string is the primary genre for that film.

```python
subgenres = final_table['genres'].str.split(',', expand = True)

subgenres
```

```python
final_genres = final_table.join(subgenres, how='inner')

final_genres
```

```python
final_genres.columns = ['movie','genres','production_budget','domestic_gross','worldwide_gross','roi','primary_genre','subgenre1','subgenre2']

final_genres
```

Now, that we have our genres broken apart, we can see what the correlations are between genre and return on investment (roi). This will only include movies that spent at least $32 million on their production budget, since that's the money we want to spend on a film as well.

We are picking roi because it not only tells us what the most profitable film is, but it is a proxy measurement for popularity. After all, revenue is from movie ticket sales, and higher sales generally mean more people liked and wanted to see it. 

```python
genre_query = """

SELECT primary_genre, roi
FROM final_genres
WHERE production_budget > 32000000
GROUP BY primary_genre
ORDER BY roi DESC

"""

genre_result = pysqldf(genre_query)

genre_result
```

It looks like movies about or with music have the highest return on investment overall. However, it looks like the term was double counted in some way, with the exact same roi on similar phrases. I'm going to keep only one term to eliminate confusion.

```python
genre_result[1:11].plot('primary_genre','roi', kind='bar')
```

```python
genre_chart = genre_result[1:11]

genre_chart
```

And now we can make a much better chart to illustrate our point.

```python
x = genre_chart['primary_genre'].astype(str)
y = genre_chart['roi'].astype(float)

plt.barh(x,y)
plt.title('ROI by Movie Genre')
plt.xlabel('Return on Investment')
plt.ylabel('Movie Genre')
```

However, genre of film doesn't necessarily guarantee a positive outcome. One thing Microsoft can do is hire a famous actor. Intuitively, we know that popular actors help sell movies. So, we can look at some data from The Numbers to determine which actor(s) Microsoft should hire.

The following is a chart that gives information on actors, the total domestic revenue of movies they starred in, and the number of movies in which they plated the lead role. The last column is an average, calculated by domestic box office divided by the number of movies.

Data for this chart was manually copied and pasted, then saved as a csv from https://www.the-numbers.com/box-office-star-records/domestic/lifetime-acting/top-grossing-leading-stars

```python
actor_rev = pd.read_csv('data/actors-by-domestic-box-office.csv')

actor_rev
```

```python
actor_rev.info()
```

Now, all we have to do is rank the actors by their average. However, we want to make sure that their average is due to repeated success, not just one movie that made a ton of money. So we will select actors who have had the lead role in at least 10 movies, so we know they have an established career.

```python
actors = """ 
SELECT *
FROM actor_rev
WHERE Movies > 10
ORDER BY Average DESC
"""

actor_results = pysqldf(actors)

actor_results
```

Sadly, we do have to remove Chadwick Boseman from the list of actors to choose from. RIP.

```python
actor_options = actor_results.drop(0)

actor_top10 = actor_options[0:10]

actor_top10
```

And now we'll make a graph of the top 10 options.

```python
actorx = actor_top10['Name']
actory = actor_top10['Average']

plt.barh(actorx,actory)
plt.title('Actors Whose Movies Make the Most Money on Average')
plt.xlabel('Average Domestic Revenue in Hundreds of Millions')
plt.ylabel('Actor/Actress')
```

## Conclusion

To recap, our data is telling us to spend at least $32 million on a production budget, make a musical, and hire Sebastian Stan as our lead actor.

However, this data can't account for the taste required to make casting decisions. Yes, Sebastian Stan is a good  actor who is in successful movies, but would he be a good fit for a musical? Would it be better instead to find a better pairing, like to have Jeremy Renner star in a sci-fi film? That may not be the most financailly successful movie according to the numbers, but that's the creative decision that could drive a movie's success. 

Additionally, the list of top actors is dominated by those who were involved in major franchises, such as Marvel or Harry Potter. Nevertheless, these actors are household names and are generally beloved by viewers, so even if their financial indicators are due to exisiting or popular IP, they are liked household names that people will want to see nonetheless.

While there is some creative discretion to be used, the data above still points us in the right direction for making a successful first movie. By spending at least the average for the production budget, Microsoft can not only likely make more money in return, but they can ensure the quality of their product is on par with those of their competitors like Sony. In fact, Microsoft might be incentivized to spend much more than $32 million to ensure that their first movie makes a strong first impression on viewers.
